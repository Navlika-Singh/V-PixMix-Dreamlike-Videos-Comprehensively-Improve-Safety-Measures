```python
df_cifar10 = pd.read_csv("cifar10_snapshots/cifar10_wrn_training_log.csv")
df_cifar10 = df_cifar10.dropna()
```

```python
plt.plot(df_cifar10["epoch"].values, df_cifar10["train_loss"].values, label="train")
plt.plot(df_cifar10["epoch"].values, df_cifar10["test_loss"].values, label="test")
plt.title("[CIFAR-10] Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
```

```python
plt.plot(df_cifar10["epoch"].values, df_cifar10["test_error(%)"].values)
plt.title("[CIFAR-10] Test Rate (%)")
plt.xlabel("Epochs")
plt.ylabel("test error (in %)")
plt.show()
```
